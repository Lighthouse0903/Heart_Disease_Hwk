```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
heart_disease = fetch_ucirepo(id=45) 
  
# data (as pandas dataframes) 
X = heart_disease.data.features
y = heart_disease.data.targets
```

```python
import pandas as pd
```

```python
X
```

```python
y
```

```python
heart_disease_df = pd.concat([X, y], axis=1)
heart_disease_df
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# plt.figure(figsize=(10, 6))
sns.scatterplot(data=heart_disease_df, x='age', y='num')
plt.title('Relationship between Age and Num')
plt.xlabel('Age')
plt.ylabel('Num')
plt.show()
```

```python
sns.scatterplot(data=heart_disease_df, x='trestbps', y='num')
plt.title('Resting Blood Pressure vs Heart Disease')
```

```python
sns.countplot(data=heart_disease_df, x='cp', hue='num')
plt.title('Chest Pain Type and Heart Disease')
```

```python
sns.countplot(data=heart_disease_df, x='num', hue='sex')
plt.title('Chest Pain Type and Heart Disease')
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=heart_disease_df, x='cp', hue='num')
plt.title('Chest Pain Type (CP) and Heart Disease')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.legend(title='Heart Disease')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=heart_disease_df, x='sex', hue='num')
plt.title('Heart Disease by Gender')
plt.xlabel('Gender (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.legend(title='Heart Disease')
plt.show()
```

```python
sns.scatterplot(data=heart_disease_df, x='age', y='chol', hue='num', palette='coolwarm')
plt.title('Age vs Cholesterol by Heart Disease')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.show()
```

```python
sns.boxplot(data=heart_disease_df, x='num', y='chol')
plt.title('Cholesterol Levels by Heart Disease')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Cholesterol')
plt.show()
```

```python
sns.stripplot(data=heart_disease_df, x='num', y='thalach', jitter=True, palette='Set2')
plt.title('Max Heart Rate by Heart Disease')
plt.xlabel('Heart Disease')
plt.ylabel('Max Heart Rate (Thalach)')
plt.show()
```

```python
plt.scatter(heart_disease_df['age'],heart_disease_df['trestbps'], s=30, c = '#b6eb7a', edgecolor = 'green', linewidth = 1, alpha = 0.8)
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure')
plt.title('Age vs RBP')
```

```python
sns.violinplot(data=heart_disease_df, x='num', y='thalach')
plt.title('Max Heart Rate by Heart Disease')
plt.xlabel('Heart Disease')
plt.ylabel('Max Heart Rate (Thalach)')
plt.show()
```

```python
sns.swarmplot(data=heart_disease_df, x='num', y='age', hue='sex', palette='Set1')
plt.title('Age Distribution by Heart Disease and Gender')
plt.xlabel('Heart Disease')
plt.ylabel('Age')
plt.legend(title='Gender')
plt.show()
```

```python
df_sorted = heart_disease_df.sort_values('age')
plt.step(df_sorted['age'], df_sorted['trestbps'], where='mid', color='purple')
plt.title('Resting Blood Pressure by Age')
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure (Trestbps)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(heart_disease_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Heart Disease Dataset')
plt.show()
```
